```python
import matplotlib.pyplot as graph
```

```python
x=[1,3,5,7,9]
y=[2,4,6,8,10]
```

```python
graph.plot(x,y)  #.plot is used to plot the points
graph.show()  #.show is used to display the graph
```

```python
graph.plot(x,y)
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y, marker = 'o')
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y,'--') # -- is used to convert the line into dotted line
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y,'*')
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y,'x')
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y,'3')
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y,'>')
graph.ylabel('Price') #.ylabel is used to define the Y Axis
graph.xlabel('Supply') #.xlabel is used to define the X Asis
graph.show()
```

```python
graph.plot(x,y,'r')# r is used to change colour
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.show()
```

```python
graph.plot(x,y,'b')
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.show()
```

```python
graph.plot(x,y,'g') 
graph.ylabel('Price') 
graph.xlabel('Supply') 
graph.show()
```

```python
graph.plot(x,y,'o:r') # o:r is used to add dots as well as change colour of the graph
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.show()
```

```python
graph.plot(x,y,'o:b', ms = 20) # ms is used to change the size of the dots
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.show()
```

```python
graph.plot(x,y,'o', ms = 20, mec = 'r') # mec is used to change the border colour of the dots
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.show()
```

```python
graph.plot(x,y,'o', ms = 20, mec = 'g', mfc = 'y')
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[42,45,60,43,38,36,48,50]
z=[100,150,120,110,101,145,136,119]
graph.plot(x,y)
graph.plot(x,z)
graph.show()
```

```python
graph.plot(x,y, 'o:r', ms=10, mec='y')
graph.plot(x,z, marker='o', ms=10, mec='b')
graph.ylabel('Price', fontdict = {'family':'serif','color':'blue','size':15}) 
graph.xlabel('Supply', fontdict = {'family':'serif','color':'green','size':15})
graph.show()
```

```python
graph.plot(x,y,'o:r') 
graph.ylabel('Price') 
graph.xlabel('Supply')
graph.title("Supply Analysis") #.title is used to add a Main Title to the graph.
graph.show()
```
