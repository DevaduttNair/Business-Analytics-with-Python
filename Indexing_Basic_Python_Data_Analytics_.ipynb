```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[1,5,7,2,3,4,7,5,3]
```

```python
plt.plot(x,y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.scatter(x,y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.scatter(x,y, marker="*")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```

```python
plt.plot(x,y, marker="D",ms=10,mec="r",mfc="g") #ms=size of marker, mec=color of border of marker, mfc=color of interior of marker.
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales vs Days")
```
