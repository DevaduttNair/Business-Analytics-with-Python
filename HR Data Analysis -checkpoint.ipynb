```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("HRDataset1.csv")
df
```

```python
# Q1. Details of Employees working as Area Sales Manager?
```

```python
df.loc[df['Position'] == "Area Sales Manager"]
```

```python
# Q2. Details of employees whose performance scores exceed the criteria?
```

```python
df.loc[df['PerformanceScore'] == "Exceeds"]
```

```python
# Q3. Names of employees whose with 4 or more special projects completed?
```

```python
for k in df.index:
    if df.loc[k,'SpecialProjectsCount'] >= 4:
        print(df.loc[k,'Employee_Name']) 
```

```python
# Q4. Names and State of Divorced & White US Citizens?
```

```python
for k in df.index:
    if df.loc[k,'MaritalDesc'] == "Divorced":
        if df.loc[k,'RaceDesc'] == "White":
            if df.loc[k,'CitizenDesc'] == "US Citizen":
                print(df.loc[k, ['Employee_Name', 'State']])
```

```python
# Q5. list of Employees who need a performance improovement plan (PIP)?
```

```python
df.loc[df['PerformanceScore'] == "PIP"]
```

```python
# Q6. Details of people working in Sales under Manager Lynn Daneault?
```

```python
df.loc[(df["Department"]=="Sales") & (df["ManagerName"]=="Lynn Daneault") ]
```

```python
# Q7. Relation between Salary & Employee Satisfaction?
```

```python
plt.scatter(df["PayRate"],df["EmpSatisfaction"])
plt.xlabel('PayRate')
plt.ylabel('Satisfaction')
```

```python
# Q8. Name of people working under John Smith whose Performance Score Fully Meets the criteria?
```

```python
for k in df.index :
    if  (df.loc[k,'ManagerName'] == "John Smith"):
        if (df.loc[k,'PerformanceScore'] == "Fully Meets"):
            print(df.loc[k,'Employee_Name'])
```

```python
# Q9. Position with Maximum Pay Rate?
```

```python
for k in df.index :
    if  (df.loc[k,'PayRate']== df['PayRate'].max()) :
        print(df.loc[k,'Position'])
```

```python
# Q10. Details of employees who fully meets the Performance Score and was terminated for a cause?
```

```python
df.loc[(df["EmploymentStatus"]=="Terminated for Cause") & (df["PerformanceScore"]=="Fully Meets") ] 
```
