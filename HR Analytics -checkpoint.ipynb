```python
for k in df.index:
    if df.loc[k,'EmploymentStatus'] == "Active":
        if df.loc[k,'Salary'] < 60000:
            print(df.loc[k,'Employee_Name']) #Finding active employees with Salary less than $60000
```

```python
for k in df.index:
    if df.loc[k,'Position'] == "Production Technician II":
        if df.loc[k,'EngagementSurvey'] == df['EngagementSurvey'].max():
            print(df.loc[k,'Employee_Name']) # Employees at Production Techincian II and highest Enagagement
```

```python
df['Salary'].max()
```

```python
# Positions drawing the Maximum salary
for k in df.index :
    if  (df.loc[k,'Salary']== df['Salary'].max()) :
        print(df.loc[k,'Position'])
```

```python
# People with below average salary
for k in df.index :
    if  (df.loc[k,'Salary'] < df['Salary'].mean()) :
        print(df.loc[k,'Employee_Name'])
```

```python
# People with below average salary and not satisfied
for k in df.index :
    if  (df.loc[k,'Salary'] < df['Salary'].mean()) :
        if (df.loc[k,'EmpSatisfaction']<3):
            print(df.loc[k,'Employee_Name'])
```

```python
# Relation betwwen Distance from home & Employee Satisfaction
plt.scatter(df["DistanceFromHome"],df["EmpSatisfaction"])
plt.xlabel('Distance')
plt.ylabel('Satisfaction')
```
